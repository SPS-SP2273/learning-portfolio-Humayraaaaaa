## 2.4 Adding another list

```python
py_list_1 = [1, 2, 3, 4, 5]
py_list_2 = [10, 20, 30, 40, 50]

np_array_1 = np.array(py_list_1)
np_array_2 = np.array(py_list_2)
```

1 to the power of 10, 2 to the power of 20 and so on

5 to the power of 50 is such a big number, that it overflows, goes around and shows a negative number

```python
py_list_1 + py_list_2
np_array_1 + np_array_2
```
