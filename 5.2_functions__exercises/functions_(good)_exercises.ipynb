```python
def convert_celsius(temperature_celsius, target_scale='Fahrenheit'):
    if target_scale == 'Kelvin':
        return temperature_celsius + 273.15
    else: 
        return(temperature_celsius * 9.5) + 32
```

```python
print(convert_celsius(20))                      
print(convert_celsius(20, target_scale='Kelvin'))
```

```python
def convert_fahrenheit(temperature_fahrenheit, target_scale='Celsius'):
    if target_scale == 'Kelvin':
        return (temperature_fahrenheit -32) * 5/9 + 273.15
    else: 
        return(temperature_fahrenheit -32 ) * 5/9
```

```python
print(convert_fahrenheit(20))                      
print(convert_fahrenheit(20, target_scale='Kelvin'))
```

```python
def convert_kelvin(temperature_kelvin, target_scale='Celsius'):
    if target_scale == 'Fahrenheit':
        return (temperature_kelvin - 273.15) * 9/5 + 32
    else: 
        return(temperature_kelvin - 273.15 )
```

```python
def convert_temperature(temperature, source_scale, target_scale): # convert target_scale to source_scale
     #Temperature in Celsius
    if source_scale == 'Celsius':
        if target_scale == 'Fahrenheit':
            return (temperature * 9/5) + 32
        elif target_scale == 'Kelvin':
            return temperature + 273.15
        else:
            return temperature 

    #Temperature in Fahrenheit
    elif source_scale == 'Fahrenheit':
        if target_scale == 'Celsius':
            return (temperature - 32) * 5/9
        elif target_scale == 'Kelvin':
            # Convert to Celsius first, then to Kelvin
            celsius = (temperature - 32) * 5/9
            return celsius + 273.15
        else:
            return temperature
    
    #Temperature in Kelvin
    elif source_scale == 'Kelvin':
        if target_scale == 'Celsius':
            return temperature - 273.15
        elif target_scale == 'Fahrenheit':
            # Convert to Celsius first, then to Fahrenheit
            celsius = temperature - 273.15
            return (celsius * 9/5) + 32
        else:
            return temperature
```

```python
print(convert_temperature(20, source_scale='Kelvin', target_scale='Celsius'))                     
print(convert_temperature(100, source_scale='Celsius', target_scale='Fahrenheit'))                     
print(convert_temperature(1, source_scale='Fahrenheit', target_scale='Kelvin'))                     
```
