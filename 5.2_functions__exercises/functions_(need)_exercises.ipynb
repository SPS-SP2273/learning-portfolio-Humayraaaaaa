```python
def greeting(name): #def is used to define the greeting
    if name == 'Batman':
        return 'Hello Batman! So, nice to meet you!'
    return f'Hello {name}!'
```

'return' is used to to get an output from the function. When Python sees a return keyword, it jumps out of the function with the return value and the function stops immediately once the return statement runs.

The return f'Hello {name}!' statement acts as an else. If name is 'Batman', it returns the message and exit the function. If the name is not Batman, it will skip the if statement and run the next line which is the second return statement. Hence, the code still works

```python
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)

    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi <= 24.9:
        return 'Normal Weight'
    elif 25 <= bmi <= 29.9:
        return 'Overweight'
    else: 
        return 'Obese'
    

```

```python
print(calculate_bmi(40, 1.6))   
print(calculate_bmi(60, 1.65))  
print(calculate_bmi(75, 1.7))  
print(calculate_bmi(95, 1.7)) 
```

```python
def calculate_factorial(n):
    if n < 0:
        return "Error: Factorial is not defined for negative numbers"
    result = 1 # 0! is 1

    for i in range(1, n+1):
        result*= i 
        
    return result
```

```python
print(f"5! is: {calculate_factorial(5)}")   
print(f"0! is: {calculate_factorial(0)}")   
print(calculate_factorial(-5))              
```
