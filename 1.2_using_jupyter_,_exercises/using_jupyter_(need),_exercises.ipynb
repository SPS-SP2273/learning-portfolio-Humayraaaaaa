The general form of a quadratic equation is:


$$
\displaystyle ax^2 +bx+c=0 
$$

Provided $a\ne0$, we can use an elementary algebraic method called completing the square to show that a quadratic equation has the following solution:

$$
x = \dfrac{-b\pm \sqrt{b^2-4ac}}{2a}
$$

### Discriminant

The quantity $\Delta$ is called the discriminant of the equation and decides the nature of its solutions. The table below shows the various possibilities.

| Discrimant| Roots |
| :---------|------:|
| $\Delta$ = $b^2-4ac =0 $ | A single solution of $-b/(2a)$|
| $\Delta$ = $b^2-4ac \gt0$ | Two distinct solutions
|  $\Delta$ = $b^2-4ac \lt0$| No real solutions; both are complex|




**What is a Git repository?**
It is a place where you can store your code and files.

Git keeps track of the **revision history of the files** in the files in our repository.

**What is a commit?**
To commit in Github means to save the project and push it up to the cloud.

**What is a push?**
To push is the action of unloading your code changes to a remote repository.

**What is a pull?**
To pull is to bring changes from a remote repository into a local environment or another branch.

**What is the difference between GitHub and GitHub Desktop?**
GitHub is a cloud-based web platform for hosting, sharing, and managing Git repositories, acting as a remote repository for collaboration. GitHub Desktop is a local GUI application designed to manage these repositories on a computer without using command-line tools.
