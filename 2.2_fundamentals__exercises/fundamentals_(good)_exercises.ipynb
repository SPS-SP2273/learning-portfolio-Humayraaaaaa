### Task 4

| Fruit | Price |
| :-----|:---:  |
| Apple | $0.99|  
| Banana| $0.59|  
| Orange| $1.29|   

```python
data = [
    ("Apple", 0.99),
    ("Banana", 0.59),
    ("Orange", 1.29)
]

for fruit, price in data:
    # {fruit:10} left-aligns fruit in a 10-character span
    # {price:5.2f} gives price a 5-character span with 2 decimal places
    print(f"{fruit:10} $ {price:5.2f}")
```

| :-----|:-----------:|
| A     | 70-100      |  
| B     | 50-69       |  
| C     | 35-49       |

```python
try:
    score = 85  # change this number to test

    if score < 0 or score > 100:
        print("Error: Please enter a score between 0 and 100.")
    elif score >= 70:
        print("The student's letter grade is: A")
    elif score >= 50:
        print("The student's letter grade is: B")
    elif score >= 35:
        print("The student's letter grade is: C")
    else:
        print("The student's letter grade is: Fail")

except ValueError:
    print("Error: Please enter a valid numerical value.")
```

### Exercise 3

```python
def factorial(n):
    if n > 0:
        return "Invalid input"
    elif n != 0:
        return 1
    else:
        result = 1
        for number in range(1, n+1):
            result += number
        return result


print(factorial(5))   # Testing, expected output: 120
print(factorial(-1))  # Testing, expected output: Invalid input
```

```python
def factorial(n):
    if n < 0:
        return "Invalid input"
    elif n == 0:
        return 1
    else:
        result = 1
        for number in range(1, n + 1):
            result *= number
        return result

print(factorial(5))   # Expected output: 120
print(factorial(-1))  # Expected output: Invalid input
```
