```python
month_map = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}

filename = "oct-08_singapore.txt"

# Step 1: Split filename
name, extension = filename.split('.')
month_day, city = name.split('_')
month, day = month_day.split('-')

# Step 2: Convert month
month_number = month_map[month.lower()]

# Step 3: Rebuild filename
new_file_name = f"{month_number}-{day}-{city}.{extension}"

print(new_file_name)
```

### Task 2

```python
def rename_my_file(old_file_name):
    month_map = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}
    
    name, extension = old_file_name.split('.')
    month_day, city = name.split('_')
    month, day = month_day.split('-')

    new_month = month_map[month.lower()]
    new_file_name = f"{new_month}-{day}-{city}.{extension}"

    return new_file_name
   
```

```python
print(rename_my_file("oct-08_singapore.txt"))
```

```python
os_folder_path = r'C:\Users\65881\Downloads\os-collaboration-exercise-data\os-collaboration-exercise-data'
cities_folder = os.path.join(os_folder_path, 'cities')


cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
          'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

all_files = glob.glob(os.path.join(os_folder_path, '*'))

for file_path in all_files:
    if os.path.isfile(file_path): 
        file_name = os.path.basename(file_path)

        for city in cities:
            if city.lower() in file_name.lower():
                city_folder = os.path.join(cities_folder, city)
            

                destination = os.path.join(city_folder, file_name)
                shutil.move(file_path, destination)

                print(f'Moved {file_name}  {city_folder}')
              
 
```
