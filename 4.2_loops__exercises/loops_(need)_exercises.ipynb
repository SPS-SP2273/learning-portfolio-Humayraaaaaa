### Exercise 1

```python
temperatures_celsius = [
    0, 5, 10, 15, 20, 25,
    30, 35, 40, 45, 50
]
for celsius in temperatures_celsius:
    fahrenheit = (celsius * 9/5) + 32
    print(f"{celsius}, {fahrenheit}")
```

### Exercise 2

```python
for letter in ['A', 'B', 'C']:
    for number in [1, 2, 3]:
        print(f'{letter}{number}', end='\t')
    print('\n')
```

```python
for i in range(1, 6):

    print(f"{i} :", end=" ")
    
    for j in range(1, 6):
        print(f"{i * j}", end="\t")
    
    print()
```
