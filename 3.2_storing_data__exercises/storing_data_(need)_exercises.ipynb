```python
import numpy as np

def pearson_correlation(x_1, x_2):
    x_1 = np.array(x_1)
    x_2 = np.array(x_2)

    mean_x_1 = np.mean(x_1)
    mean_x_2 = np.mean(x_2)

    dev_x_1 = x_1 - mean_x_1
    dev_x_2 = x_2 - mean_x_2

    numerator = np.sum(dev_x_1 * dev_x_2)

    sum_sq_dev_x_1 = np.sum(dev_x_1**2)
    sum_sq_dev_x_2 = np.sum(dev_x_2**2)
    denominator = np.sqrt(sum_sq_dev_x_1 * sum_sq_dev_x_2)

    if denominator == 0:
        return 0

    r = numerator / denominator
    return r


x_1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
x_2 = [1, 8, 27, 64, 125, 216, 343, 512, 729, 1000]

correlation_coefficient = pearson_correlation(x_1, x_2)
print(f"The Pearson correlation coefficient is: {correlation_coefficient}")
```
